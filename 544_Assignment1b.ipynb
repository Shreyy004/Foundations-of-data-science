```python
import pandas as pd

url = "https://github.com/flaviovdf/evcomp2018/raw/master/data/nc-est2015-agesex-res.csv"
df = pd.read_csv(url)
```

```python
df.rename(columns={"POPESTIMATE2010": "2010", "POPESTIMATE2014": "2014"}, inplace=True)
```

```python
df['Change'] = df['2014'] - df['2010']
df['Percentage Change'] = (df['Change'] / df['2010']) * 100
df.head()
```

```python
df[(df['AGE'] >= 67) & (df['AGE'] <= 71)]
```

```python
infants = df[df['AGE'] == 0]
gender_ratio_infants = infants[infants['SEX'] == 1]['2014'].values[0] / infants[infants['SEX'] == 0]['2014'].values[0]
print(f'Gender ratio of boys to girls among infants (age = 0): {gender_ratio_infants}')
```

```python
infants=df[df['AGE']==0]
males=infants[infants['SEX']==0]['2014'].sum()
females=infants[infants['SEX']==1]['2014'].sum()
df['gender_ratio']=males/females
df.head()
```

```python
years_with_ratio=df[(df['gender_ratio']>=2)&(df['gender_ratio']<=4)]
print(years_with_ratio)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.plot(df['AGE'],df['gender_ratio'],linestyle='-',color='b')
plt.xlabel('Age')
plt.ylabel('Gender Ratio(Females/Males)')
plt.title('Gender Ratio by Age')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
url3 = 'https://github.com/flaviovdf/evcomp2018/raw/master/data/educ_inc.csv'
df2 = pd.read_csv(url3)

df2.head()
```

```python
df2.shape
```

```python
df2.dtypes
```

```python
df2['Year'] = pd.to_datetime(df2['Year'], errors='coerce')
df2['Year'] = df2['Year'].dt.year
df2.head()
```

```python
def extract_end_age(age_range):
    ages = ''.join([c if c.isdigit() else ' ' for c in age_range]).split()
    if len(ages) == 2:
        end_age = int(ages[1])
        return end_age
    return None

df2['Age'] = df2['Age'].apply(extract_end_age) 
```

```python
df2['Age']=pd.to_numeric(df2['Age'],errors='coerce')
```

```python
df2[(df2['Year'] == 2014) & (df2['Age'] > 17)]
```

```python
contingency_table=pd.pivot_table(df,values='Population Count',index='Educational Attainment',columns='Personal Income',aggfunc='sum')
contingency_table_percentage=contingency_table.div(contingency_table.sum(axis=1),axis=0)*100
print(contingency_table_percentage)
```

```python
no_high_school_label = 'No high school diploma'
bachelor_or_higher_label = 'Bachelor\'s degree or higher'

if no_high_school_label in contingency_table_percentage.index and bachelor_or_higher_label in contingency_table_percentage.index:
    no_high_school = contingency_table_percentage.loc[no_high_school_label]
    bachelor_or_higher = contingency_table_percentage.loc[bachelor_or_higher_label]

    combined_df = pd.DataFrame({
        'No High School': no_high_school,
        'Bachelor\'s Degree or Higher': bachelor_or_higher
    })

    combined_df = combined_df.sort_values(by='No High School', ascending=False)

    # Plot
    plt.figure(figsize=(12, 8))

    bar_height = 0.4
    index = range(len(combined_df))

    plt.barh(index, combined_df['Bachelor\'s Degree or Higher'], bar_height, alpha=0.7, label=bachelor_or_higher_label)
    
    plt.barh([i + bar_height for i in index], combined_df['No High School'], bar_height, alpha=0.7, label=no_high_school_label,color='orange')

    plt.gca().invert_yaxis()

    plt.ylabel('Personal Income')
    plt.xlabel('Percentage')
    plt.yticks([i + bar_height / 2 for i in index], combined_df.index)  # Set y-ticks to the middle of the grouped bars
    plt.title('Personal Income Distribution: No High School Diploma vs Bachelor\'s Degree or Higher')
    plt.legend()
    plt.show()
else:
    print("One or both of the specified educational attainment levels are not in the dataset.")
```
