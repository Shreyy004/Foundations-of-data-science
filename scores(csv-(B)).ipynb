```python
data={
    "Ajay":[95,80,99,98,97],
    'Aman': [67,45,39,45,35], 
    'Abi':[99,89,96,93,95],
    'Joy': [33,26,41, 37,32]
}
subjects = ['Maths', 'Science', 'English', 'History', 'Geography']
scores = pd.DataFrame(data, index=subjects)
print(scores)
```

```python
scores.corr()
```

```python
scores.cov()
```

```python
five_number_summary = scores.describe().loc[['min', '25%', '50%', '75%', 'max']]
five_number_summary
```

```python
# Display marks of Ajay, Aman, and Joy in Maths using iloc
maths_scores = scores.iloc[0, [0, 1, 3]]  # 0th row for Maths and columns for Ajay, Aman, Joy
print("Marks in Maths:")
print(maths_scores)
```

```python
scores["Ajay"].dtype
```

```python
import matplotlib.pyplot as plt

# Bar chart of the scores
scores.plot(kind='bar')
plt.ylabel('Marks')
plt.title('Marks scored by students')
plt.show()
```

```python
# Box plot of the scores
scores.plot(kind='box')
plt.title('Box Plot of Marks')
plt.show()
```

```python
# Mean and Median scores
mean_scores = scores.mean()
median_scores = scores.median()
print("Mean scores:")
print(mean_scores)
print("Median scores:")
print(median_scores)
```

```python
# Create a new dataframe with NaN values
df_with_nan = pd.DataFrame([[1, np.nan, 2], [2, 3, 4], [np.nan, 4, 5]])

# Replace NA values with median
df_filled = df_with_nan.fillna(df_with_nan.median())
print("Dataframe with NA replaced by median:")
print(df_filled)
```

```python
# Drop Ajay and Joy from dataframe
new_scores = scores.drop(columns=['Ajay', 'Joy'])
print("Dataframe after dropping Ajay and Joy:")
print(new_scores)
```
