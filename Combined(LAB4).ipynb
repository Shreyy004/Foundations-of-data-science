```python
Mean=Sud["Age"].mean()
Sud["Age"].fillna(Mean , inplace= True)
Sud
```

```python
Sud["Embarked"]=Sud["Embarked"].fillna(Sud["Embarked"].mode()[0])
```

```python
Medi=df["Fare"].median()
for x in Sud.index:
    if df.loc[x , "Fare"]<=0:
        df.loc[x , "Fare"]= Medi
print("Number of negative fares:", (df['Fare'] < 0).sum()) 
```

```python
mean=Sud["Age"].mean()
for x in Sud.index:
    if Sud.loc[x,"Age"]>100:
        Sud.loc[x,"Age"]=mean
print("Number of ages greater than 100:", (Sud['Age'] > 100).sum())
```

```python
Sud["Fare"].astype(float)
```

```python
Sud["Cabin"] = Sud["Cabin"].str.split().str[0]
```

```python
print(Sud["Cabin"].head())
```

```python
def extract_first_cabin(cabin):
    if pd.isnull(cabin):
        return cabin  # Return NaN if the cabin value is missing (null)
    # Split by spaces and return the first cabin number
    return cabin.split()[0]

# Apply the function to the 'Cabin' column
Sud["firstCabin"] = Sud["Cabin"].apply(extract_first_cabin)
Sud
```

```python
Gl=pd.read_csv("global_power_plant_database.csv")
Gl
```

```python
Gl.isnull().sum()
```

```python
# Identify erroneous values (outside the range 1900-2024)
erroneous_values = Gl[(Gl["commissioning_year"] < 1900) | (Gl["commissioning_year"] > 2024)]
print(erroneous_values["commissioning_year"])
```

```python
# Replace erroneous values with the median year
median_year = Gl["commissioning_year"].median()
Gl.loc[(Gl["commissioning_year"] < 1900) | (Gl["commissioning_year"] > 2024), "commissioning_year"] = median_year
```

```python
# Check if there are any values still outside the range
print(Gl[(Gl["commissioning_year"] < 1900) | (Gl["commissioning_year"] > 2024)])
```

```python
if Gl['capacity_mw'].dtype != 'float64':
    Gl['capacity_mw'].astype(float)
print("\nAfter conversion:")
print(Gl.dtypes)
```
